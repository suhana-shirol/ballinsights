```python
import pandas as pd
    
agg = pd.DataFrame()

teams_df = pd.read_csv('MTeams.csv')
teams_df = teams_df.loc[teams_df['LastD1Season'] == 2022]
teams_df = teams_df[['TeamID', 'TeamName']]

MMteams = pd.read_csv('MNCAATourneySeeds.csv')
MMteams = MMteams.loc[MMteams['Season'] == 2021][['TeamID']]

agg1 = MMteams.set_index('TeamID').join(teams_df.set_index('TeamID'))
display(agg1)

stats_df = pd.read_csv('MNCAATourneyDetailedResults.csv')
stats_df = stats_df.loc[stats_df['Season'] == 2021]
display(stats_df)

win = stats_df[['WTeamID', 'WScore', 'WLoc']].copy()
win['WFG%'] = stats_df['WFGM']/stats_df['WFGA'] * 100
win['W3Pt%'] = stats_df['WFGM3']/stats_df['WFGA3'] * 100
win['W2Pt%'] = (stats_df['WFGM']-stats_df['WFGM3'])/(stats_df['WFGA']-stats_df['WFGA3'])
win['WFT%'] = stats_df['WFTM']/stats_df['WFTA'] * 100
win['WAssists'] = stats_df['WAst']
display(win.sort_values(by=['WTeamID']))
win = win.groupby(['WTeamID']).mean()
display(win)

lose = stats_df[['LTeamID', 'LScore']].copy()
lose['LFG%'] = stats_df['LFGM']/stats_df['LFGA'] * 100
lose['L3Pt%'] = stats_df['LFGM3']/stats_df['LFGA3'] * 100
lose['L2Pt%'] = (stats_df['LFGM']-stats_df['LFGM3'])/(stats_df['LFGA']-stats_df['LFGA3'])
lose['LFT%'] = stats_df['LFTM']/stats_df['LFTA'] * 100
lose['LAssists'] = stats_df['LAst']
lose = lose.groupby(['LTeamID']).mean()
display(lose)

agg2 = lose.join(win)
tourney_stats = agg1.join(agg2)
tourney_stats.to_csv('tourney_stats.csv')
```

```python
regstats_df = pd.read_csv('MRegularSeasonDetailedResults.csv')
regstats_df = regstats_df.loc[regstats_df['Season'] == 2021]
display(regstats_df)

regwin = regstats_df[['WTeamID', 'WScore', 'WLoc']].copy()
regwin['WFG%'] = regstats_df['WFGM']/regstats_df['WFGA'] * 100
regwin['W3Pt%'] = regstats_df['WFGM3']/regstats_df['WFGA3'] * 100
regwin['W2Pt%'] = (regstats_df['WFGM']-regstats_df['WFGM3'])/(regstats_df['WFGA']-regstats_df['WFGA3'])
regwin['WFT%'] = regstats_df['WFTM']/regstats_df['WFTA'] * 100
regwin['WAssists'] = regstats_df['WAst']
display(regwin.sort_values(by=['WTeamID']))
regwin = regwin.groupby(['WTeamID']).mean()
display(regwin)

rlose = regstats_df[['LTeamID', 'LScore']].copy()
rlose['LFG%'] = regstats_df['LFGM']/regstats_df['LFGA'] * 100
rlose['L3Pt%'] = regstats_df['LFGM3']/regstats_df['LFGA3'] * 100
rlose['L2Pt%'] = (regstats_df['LFGM']-regstats_df['LFGM3'])/(regstats_df['LFGA']-regstats_df['LFGA3'])
rlose['LFT%'] = regstats_df['LFTM']/regstats_df['LFTA'] * 100
rlose['LAssists'] = regstats_df['LAst']
rlose = rlose.groupby(['LTeamID']).mean()
display(rlose)

agg3 = rlose.join(regwin)
regszn_stats = agg1.join(agg3)
regszn_stats.to_csv('regszn_stats.csv')
```
