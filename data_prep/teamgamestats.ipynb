```python
import pandas as pd
    
agg = pd.DataFrame()

teams_df = pd.read_csv('MTeams.csv')
teams_df = teams_df.loc[teams_df['LastD1Season'] == 2022]
teams_df = teams_df[['TeamID', 'TeamName']]

city_df = pd.read_csv('Cities.csv')
##display(city_df)

mmloc_df = pd.read_csv('MGameCities.csv')
mmloc_df = mmloc_df.loc[mmloc_df['CRType'] == 'NCAA']
mmloc_df = mmloc_df.loc[mmloc_df['Season'] == 2021]
daynum = mmloc_df['DayNum'].astype(str)
wteamid = mmloc_df['WTeamID'].astype(str)
mmloc_df['GameID'] = daynum + wteamid
##display(mmloc_df)

city_agg = mmloc_df.set_index('CityID').join(city_df.set_index('CityID'))
city_agg = city_agg[['GameID', 'DayNum', 'WTeamID', 'LTeamID', 'City', 'State']]
city_agg.reset_index(inplace=True)
##city_agg = city_agg.set_index('GameID')
##display(city_agg)

MMteams = pd.read_csv('MNCAATourneySeeds.csv')
MMteams = MMteams.loc[MMteams['Season'] == 2021][['TeamID']]
##display(MMteams)

agg1 = MMteams.set_index('TeamID').join(teams_df.set_index('TeamID'))
agg1.reset_index(inplace=True)
##display(agg1)

stats_df = pd.read_csv('MNCAATourneyDetailedResults.csv')
stats_df = stats_df.loc[stats_df['Season'] == 2021]
daynum = stats_df['DayNum'].astype(str)
wteamid = stats_df['WTeamID'].astype(str)
stats_df['GameID'] = daynum + wteamid
##display(stats_df.sort_values(by=['WTeamID']))

agg = stats_df[['WTeamID', 'GameID', 'WLoc', 'WScore', 'WAst', 'WOR', 'WDR', 'WTO', 'WStl', 'WBlk']].copy()
agg['W3Pt%'] = stats_df['WFGM3']/stats_df['WFGA3']
agg['WFT%'] = stats_df['WFTM']/stats_df['WFTA']
##display(agg.sort_values(by=['WTeamID']))

citycomplete = pd.merge(agg, city_agg[['GameID', 'CityID', 'City', 'State']], on='GameID')
cityteamcomplete = citycomplete.merge(agg1[['TeamID','TeamName']], how = 'left', left_on = 'WTeamID', right_on = 'TeamID').drop(columns=['TeamID'])
cityteamcomplete.sort_values(by=['WTeamID']).set_index('WTeamID').to_csv('mmteamgamestats.csv')
```

```python
regloc_df = pd.read_csv('MGameCities.csv')
regloc_df = regloc_df.loc[regloc_df['CRType'] == 'Regular']
regloc_df = regloc_df.loc[regloc_df['Season'] == 2021]
daynum = regloc_df['DayNum'].astype(str)
wteamid = regloc_df['WTeamID'].astype(str)
regloc_df['GameID'] = daynum + wteamid
##display(regloc_df)

regcity_agg = regloc_df.set_index('CityID').join(city_df.set_index('CityID'))
regcity_agg = regcity_agg[['GameID', 'DayNum', 'WTeamID', 'LTeamID', 'City', 'State']]
regcity_agg.reset_index(inplace=True)
##city_agg = city_agg.set_index('GameID')
##display(regcity_agg)

regstats_df = pd.read_csv('MRegularSeasonDetailedResults.csv')
regstats_df = regstats_df.loc[regstats_df['Season'] == 2021]
regdaynum = regstats_df['DayNum'].astype(str)
regwteamid = regstats_df['WTeamID'].astype(str)
regstats_df['GameID'] = regdaynum + regwteamid
##display(regstats_df.sort_values(by=['WTeamID']))

regagg = regstats_df[['WTeamID', 'GameID', 'WLoc', 'WScore', 'WAst', 'WOR', 'WDR', 'WTO', 'WStl', 'WBlk']].copy()
regagg['W3Pt%'] = regstats_df['WFGM3']/regstats_df['WFGA3']
regagg['WFT%'] = regstats_df['WFTM']/regstats_df['WFTA']
#display(regagg.sort_values(by=['WTeamID']))

regcitycomplete = pd.merge(regagg, regcity_agg[['GameID', 'CityID', 'City', 'State']], on='GameID')
regcityteamcomplete = regcitycomplete.merge(agg1[['TeamID','TeamName']], how = 'left', left_on = 'WTeamID', right_on = 'TeamID').drop(columns=['TeamID'])
regcityteamcomplete2 = regcityteamcomplete.dropna(subset=['TeamName'])
regcityteamcomplete2.sort_values(by=['WTeamID']).set_index('WTeamID').to_csv('regteamgamestats.csv')
```
